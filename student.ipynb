## 4.Data Analysis

```python
## 1.Removing rows with missing values
df = df.dropna()
df
```

```python

## 2.Standardizing column names
df.columns = df.columns.str.lower().str.replace(' ', '_')
df
```

```python
## 4.Removing duplicates
df = df.drop_duplicates()
df
```

## There are 1199 rows and 8 columns after cleaning missing values and duplicate values

```python
## 5.Checking remaining data
df.info()
```

```python
## 6. Dropping irrelevant columns
df = df.drop(columns=['unnamed:_0'])
df
```

```python
## 7.Fatality by location
df['fat'] = pd.to_numeric(df['fat'], errors='coerce')
fatality_by_location = df.groupby('location')['fat'].sum().sort_values(ascending=False)

fatality_by_location
```

```python
# Fatalities by location plot
plt.figure(figsize=(12, 6))
fatality_by_location.head().plot(kind='bar', color='red')
plt.title('Fatalities by Location')
plt.xlabel('Location')
plt.ylabel('Number of Fatalities')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


```

```python
## 8.Fatality Rate by aircraft type
df['fat'] = pd.to_numeric(df['fat'], errors='coerce')
```

```python
fatality_by_aircraft = df.groupby('type')['fat'].sum().sort_values()
```

```python
fatality_by_aircraft
```

```python
# Safest 5 on the left
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
fatality_by_aircraft.head(5).plot(kind='barh', color='green', ax=ax1)
ax1.set_title('Top 5 Safest Aircraft')
ax1.set_xlabel('Total Fatalities')

# Riskiest 5 on the right
fatality_by_aircraft.tail(5).plot(kind='barh', color='red', ax=ax2)
ax2.set_title('Top 5 Riskiest Aircraft')
ax2.set_xlabel('Total Fatalities')
plt.suptitle('Fatality Rate by Aircraft Type', fontsize=14)
plt.tight_layout()
plt.show()
```

## As we can see the **Dassault Falcon aircafts** and the **ATR 42-300** are the safest aircraft options for purchase they have with zero fatalities. While the **Boeing 737 series planes** and **IIyushin II-76TD** are not recommended for purchase beacuse they have very high fatalities.

## 5.Insights
- Out of 506 unique aircraft types, the Dassault Falcon series (7X, 900B, 900EX, 900EX EASy) and ATR 42-300 recorded zero fatalities.
- The Boeing 737 MAX 8 is the riskiest with 346 fatalities, followed by Ilyushin Il-76TD (258) and other Boeing 737 variants.
- Substantial damage is the most common outcome across all incidents (661 out of 1,199), meaning most accidents result in serious damage.
- Write-offs (315) are the second most common, meaning the aircraft is destroyed.

## 6. Recommendations
- The company should prioritize purchasing Dassault Falcon aircraft (7X, 900B, 900EX) as they have the lowest operational risk.
- The company should avoid Boeing 737 variants, especially the 737 MAX 8, due to the high fatality and damage record.
- The ATR 42-300 is also a safe option if the company needs smaller aircraft.
